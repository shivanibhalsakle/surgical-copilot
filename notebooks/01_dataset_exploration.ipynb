```python
summary = []

for name, df in tables.items():
    summary.append({
        "Table": name,
        "Rows": len(df),
        "Columns": len(df.columns)
    })

summary = (
    pd.DataFrame(summary)
      .sort_values("Rows", ascending=False)
)

summary
```

```python
column_summary = []

for name, df in tables.items():

    for column in df.columns:

        column_summary.append({
            "Table": name,
            "Column": column,
            "Datatype": str(df[column].dtype),
            "Missing (%)":
                round(df[column].isna().mean() * 100, 2)
        })

column_summary = pd.DataFrame(column_summary)

column_summary.head()
```

```python
column_summary.to_csv(
    "../docs/inspire_columns.csv",
    index=False
)
```

```python
def feature_coverage(df, feature_column):
    return (
        df[df["item_name"] == feature_column]
        .groupby("op_id")
        .size()
        .shape[0]
    )
```

```python
total_operations = tables["operations"]["op_id"].nunique()

total_operations
```

```python
vital_features = [
    "hr",
    "nibp_sbp",
    "nibp_dbp",
    "rr",
    "spo2",
    "bt"
]

for feature in vital_features:
    coverage = feature_coverage(
        tables["vitals"],
        feature
    )

    print(
        f"{feature:12} {coverage:7} ({coverage/total_operations:.1%})"
    )
```
